```python
import numpy as np #type:ignore
import sympy as sp #type:ignore

class Finder:
    
    def __init__(self):
        self.dimension = np.nan
        self.punto_evaluacion_derivada = np.nan
        self.resultado_derivada = np.nan
        self.punto_evaluacion_polinomio = np.nan
        self.resultado_polinomio = np.nan
    
    def pedir_datos(self):
        
        self.dimension = int(input("Ingrese la dimension del espacio de polinomios: "))
        self.punto_evaluacion_derivada = float(input("Ingrese el punto de evaluacion para la derivada: "))
        self.resultado_derivada = float(input(f"Ingrese el resultado para la evaluacion de la derivada en P({self.punto_evaluacion_derivada})"))
        self.punto_evaluacion_polinomio = float(input("Ingrese el punto de evalucion para el polinomio: "))
        self.resultado_polinomio = float(input(f"Ingrese el resultado para la evaluacion del polinomio en P({self.punto_evaluacion_polinomio})"))
        
    def generar_polinomio(self):
        
        grado = self.dimension - 1 
        x = sp.symbols('x')  # Variable del polinomio
        coeficientes = sp.symbols(f'a:{grado+1}')  # Crea coeficientes simb√≥licos a0, a1, ..., an
        polinomio = sum(coeficientes[i] * x**i for i in range(grado+1))  # Forma del polinomio
        return polinomio
        
    def hallar_derivada(self, polinomio):
        x = sp.symbols('x')
        derivada = sp.diff(polinomio, x)
        return derivada
    
    def evaluar_punto_polinomio(self, polinomio):
        x = sp.symbols('x')        
        polinomio_resultante = polinomio.evalf(subs={x:self.punto_evaluacion_polinomio})
        return polinomio_resultante
    
    def evaluar_punto_derivada(self, derivada):
        x = sp.symbols('x')
        derivada_resultante =  derivada.evalf(subs={x:self.punto_evaluacion_derivada})
        return derivada_resultante
    
    def igualar_polinomio_evaluado(self, polinomio_evaluado):
        
        polinomio_evaluado_igualado = sp.Eq(polinomio_evaluado, self.resultado_polinomio)
        return polinomio_evaluado_igualado
    
    def igualar_derivada_evaluada(self, derivada_evaluada):
        
        derivada_evaluada_igualada = sp.Eq(derivada_evaluada, self.resultado_derivada)
        return derivada_evaluada_igualada
    
    def despejar_constante_polinomio_evaluado_igualad(self, polinomio_evaluado_igualado):
        a0 =  sp.symbols('a0')
        a0_despejado = sp.solve(polinomio_evaluado_igualado, a0)[0]
        return a0_despejado
    
    def despejar_constante_derivada_evaluada_igualada(self, derivada_evaluada_igualada):
        
        a1 = sp.symbols('a1')
        a1_despejado = sp.solve(derivada_evaluada_igualada, a1)[0]
        return a1_despejado   
        
    

finder = Finder()
finder.pedir_datos()
polinomio = finder.generar_polinomio()
derivada = finder.hallar_derivada(polinomio)

polinomio_evaluado = finder.evaluar_punto_polinomio(polinomio)
derivada_evaluada = finder.evaluar_punto_derivada(derivada)

polinomio_evaluado_igualado = finder.igualar_polinomio_evaluado(polinomio_evaluado)
derivada_evaluada_igualada = finder.igualar_derivada_evaluada(derivada_evaluada)
a0_despejado = finder.despejar_constante_polinomio_evaluado_igualad(polinomio_evaluado_igualado)
a1_despejado = finder.despejar_constante_derivada_evaluada_igualada(derivada_evaluada_igualada)


print("Polinomio: ")
print(polinomio)
print(polinomio_evaluado)
print(polinomio_evaluado_igualado)
print(a0_despejado)

print("\nDerivada: ")
print(derivada)
print(derivada_evaluada)
print(derivada_evaluada_igualada)
print(a1_despejado)

        
        
        
```
